```python
import pandas as pd
import numpy as np
from datasets import Dataset
from transformers import AutoTokenizer, DataCollatorWithPadding, AutoModelForSequenceClassification, TrainingArguments, Trainer, AutoModelForSequenceClassification, TrainingArguments, Trainer
from chamd import ChatReader
from google.colab import drive
import os
import evaluate
from transformers import pipeline
drive.mount('/content/drive')
samples_df = pd.read_excel("/content/drive/MyDrive/Ex_sentences.xlsx")
```

```python
models = {}
tokenizers = {}

ep_6_models = [4,5,6,7,8,9,14,15,16,17,30,31,32,33,34,40,41,42,43,44,47,48,49]
for category in ["I" ,"J"]:
  category_results = {}
  for sample in [54,55,56]:
    sample_idx = list(np.where(samples_df["Unnamed: 0"] == sample))[0][0]
    sample_results = []
    for run in range(0,50):
       if (str(run) + "_" + str(sample)) not in models.keys():
          dir = '/content/drive/MyDrive/val27_ep6_models/' if run in ep_6_models else "/content/drive/MyDrive/val27_models/"
          try:
            models[str(run) + "_" + str(sample)] = AutoModelForSequenceClassification.from_pretrained(dir + str(run) + "_" + str(sample))
            tokenizers[str(run) + "_" + str(sample)] = AutoTokenizer.from_pretrained(dir + str(run) + "_" + str(sample))
          except Exception:
            print("Problems loading model.. pass")
            sample_results.append("NA")
            continue
       try:
        classifier = pipeline("sentiment-analysis", model=models[str(run) + "_" + str(sample)], tokenizer=tokenizers[str(run) + "_" + str(sample)])
       except Exception:
        print("Problem making prediction")
        sample_results.append("NA")
        continue
       print(samples_df[category][sample_idx])
       result = classifier(samples_df[category][sample_idx])[0]
       score = float(result["score"]) if result["label"] == "POSITIVE" else (1 - float(result['score']))
       sample_results.append(score)
    category_results[sample] = sample_results
  results = pd.DataFrame(category_results)
  results.to_csv('/content/drive/MyDrive/sample_result54_55_56' + category + ".csv")
  print("run" + str(category))
```

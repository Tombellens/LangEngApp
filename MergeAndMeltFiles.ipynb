```python
import pandas as pd
from google.colab import drive
import os
drive.mount('/content/drive')
```

```python
categories = ["original", "I", "J"]
ids = ["54_55_56", "_70_101"]
for category in categories:
  df_1 = pd.read_csv("/content/drive/MyDrive/sample_result" + ids[0] + category + ".csv")
  df_2 = pd.read_csv("/content/drive/MyDrive/sample_result" + ids[1] + category + ".csv")
  merged = pd.merge(df_1, df_2, on="Unnamed: 0")
  merged.to_csv("/content/drive/MyDrive/IJmerged" + category + ".csv")
```

```python
df_original = pd.read_csv("/content/drive/MyDrive/IJmergedoriginal.csv")
group_list = []
for i in range(0, len(df_original.index)):
  group_list.append("original")
df_original["group"] = group_list

group_list = []
for i in range(0, len(df_original.index)):
  group_list.append("sample")
for category in categories[1:]:
  df_sample = pd.read_csv("/content/drive/MyDrive/IJmerged" + category + ".csv")
  df_sample["group"] = group_list
  pd.concat([df_sample, df_original]).drop([ "Unnamed: 0.1" ,"Unnamed: 0"], axis=1).to_csv("/content/drive/MyDrive/IJoriginal_concat_" + category + ".csv")
```

```python
for category in categories[1:]:
  df = pd.read_csv("/content/drive/MyDrive/IJoriginal_concat_" + category + ".csv")
  df.melt(id_vars=["Unnamed: 0", "group"], var_name="sample", value_name="score").to_csv("/content/drive/MyDrive/IJmelted_" + category + ".csv")
```
